```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
!pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
# Select only numerical features for correlation analysis
numerical_features = df_tracks.select_dtypes(include=['number']).columns
corr_df = df_tracks[numerical_features].drop(["key", "mode", "explicit"], axis=1, errors='ignore').corr(method="pearson")

plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
df_tracks["dates"] = df_tracks.index.get_level_values('release_date')
# The error is likely due to inconsistent date formats in the 'release_date' column.
# Using the 'infer_datetime_format' and 'errors' arguments allows pandas to handle these inconsistencies.
df_tracks.dates = pd.to_datetime(df_tracks.dates, infer_datetime_format=True, errors='coerce')
years = df_tracks.dates.dt.year
```

pip install -- user seaborn == 0.11.0

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
# Calculate the total duration for each unique year
total_dr_by_year = df_tracks.groupby(years)['duration'].sum()

# Extract unique years and corresponding total durations
unique_years = total_dr_by_year.index
total_durations = total_dr_by_year.values

# Create the bar plot
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(x=unique_years, y=total_durations, ax=ax, errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
df_genre=pd.read_csv('/content/SpotifyFeatures.csv')
df_genre.head()
```

```python
plt.title("Distribution of Genre")
sns.color_palette("rocket", as_cmap=True)
# Check if the column name is 'genre' instead of 'genres'
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel("Duration in mill seconds")
plt.ylabel("Genres")
```
